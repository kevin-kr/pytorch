```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_sample_images
from sklearn.decomposition import PCA
```

```python
samples = load_sample_images().images

plt.imshow(samples[0], cmap='Greys_r')
plt.axis('off')

samples[0].shape
```

```python
sample = samples[0]
# RGB to grayscale
original = np.mean(sample, 2)
sample = np.mean(sample, 2)

pca = PCA(.9)

sample = pca.fit_transform(sample)

pca.n_components_
```

```python
approximation = pca.inverse_transform(sample)
```

```python
fig = plt.figure(figsize=(15, 10))

fig.add_subplot(1, 2, 1)
plt.imshow(approximation, cmap='Greys_r')
plt.axis('off')
plt.title('90% of explained variance')

fig.add_subplot(1, 2, 2)
plt.imshow(original, cmap='Greys_r')
plt.axis('off')
plt.title('original')
```
